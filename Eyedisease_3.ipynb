```python
from keras.models import load_model
import cv2
import numpy as np
```

```python
model = load_model('eyediseasemodel.h5')                        #saved model

eye_clsfier = cv2.CascadeClassifier('haarcascade_eye.xml')      
source = cv2.VideoCapture(0)                                       # 0 - for web cam
# img = cv2.imread('img32.jpg')                                        #by image

labels_dict = {0:'Cataracts',1:'Crossed_Eyes',2:'Glaucoma',3:'Normal',4:'Uveitis'}       
color_dict = {0:(255,0,0),1:(0,0,255),2:(255,0,0),3:(0,255,0),4:(0,0,255)}              #BGR
```

```python
while(True):
    
    ret,img = source.read()                              #read every object in the web cam
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)            #convert to gray scale
    eyes = eye_clsfier.detectMultiScale(gray,1.3,5)        #1.3-scale factor, 5 - min neighbor
    
    for (x,y,w,h) in eyes:
        
        eye_img = gray[y:y+w,x:x+h] 
        resized = cv2.resize(eye_img,(100,100))            #resize 100*100
        normalized = resized/255.0                         #reduce to 0-1
        reshaped = np.reshape(normalized,(1,100,100,1))    #reshape-(batch size-1, size- 100,100, channel- 1)
        result = model.predict(reshaped)                   #give to predict
        
        label = np.argmax(result,axis=1)[0]                  #predict which one has max intensity
        
        cv2.rectangle(img,(x,y),(x+w,y+h),color_dict[label],2)     #draw a rectangle on the eyes , 2 - thickness
        cv2.rectangle(img,(x,y-40),(x+w,y),color_dict[label],-1)
        cv2.putText(img, labels_dict[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)  #put text on the box(rectangle) # 0.8-> text size,(255)-color, 2-darkness
        
    cv2.imshow('Eye detection',img)
    key = cv2.waitKey(1)
    
    if(key==27):
        break
        
cv2.destroyAllWindows()
source.release()
# img.release()
```
